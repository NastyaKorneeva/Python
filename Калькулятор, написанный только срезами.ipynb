```python
def rasdel(drob):
    
    if '/' in drob:
        
        symv = drob.find('/')
        l = len(drob)
        znam = drob[symv+1:l:1]
        chisl = drob[0:symv:1]
        
    else:
        
        znam = 1
        chisl = drob
        
    return chisl, znam
```

```python
def smesh(drob):
    
    if '(' in drob:
        
        l = len(drob)
        symv = drob.find('(')
        cel = int(drob[0:symv:1])
        drob = drob[symv+1:l-1:1]
        
    else:
        
        cel = -1
        
    symv = drob.find('/')
    l = len(drob)
    znam = int(drob[symv+1:l:1])
    chisl = int(drob[0:symv:1])
    
    if cel == -1:
        
        return str(chisl) + '/' + str(znam)
    
    else:
    
        return str(znam * cel + chisl) + '/' + str(znam)
```

```python
def sokra(drob):
    
    minus = 0
    drob = rasdel(drob)
    chisl = drob[0]
    
    if chisl[0] == '-':
        
        minus = 12
        l = len(chisl)
        chisl = chisl[1:l:1]
        
    chisl = int(chisl)
    znam = int(drob[1])
    sznam = znam
    schisl = chisl 
    
    if chisl == 0:
        
        return 0
    
    while sznam != schisl:
        
        if sznam > schisl: 
            
            sznam = sznam - schisl
            
        else: 
            
            schisl = schisl - sznam

    znam = znam // schisl
    chisl = chisl // schisl
    
    if minus == 12:
        
        return '-' + str(chisl) + '/' + str(znam)
    
    return str(chisl) + '/' + str(znam)
```

```python
def summa(drob1, drob2):
    
    znam = 12
    drob1 = rasdel(drob1)
    drob2 = rasdel(drob2)
    chisl1 = int(drob1[0])
    znam1 = int(drob1[1])
    chisl2 = int(drob2[0])
    znam2 = int(drob2[1])
    rez = str(chisl1 * znam2 + chisl2 * znam1) + '/' + str(znam1 * znam2)
    rez = sokra(rez)
    
    if '/' in str(rez):
        
        symv = rez.find('/')
        l = len(rez)
        znam = rez[symv+1:l:1]
        chisl = rez[0:symv:1]
    
    if int(znam) == 1:
        
        return chisl
    
    return rez
```

```python
def ruznost(drob1, drob2):

    znam = 12
    drob1 = rasdel(drob1)
    drob2 = rasdel(drob2)
    chisl1 = int(drob1[0])
    znam1 = int(drob1[1])
    chisl2 = int(drob2[0])
    znam2 = int(drob2[1])
    rez = str(chisl1 * znam2 - chisl2 * znam1) + '/' + str(znam1 * znam2)
    rez = sokra(rez)
    
    if '/' in str(rez):
        
        symv = rez.find('/')
        l = len(rez)
        znam = rez[symv+1:l:1]
        chisl = rez[0:symv:1]
    
    if int(znam) == 1:
        
        return chisl
    
    return rez
```

```python
def umnoz(drob1, drob2):

    znam = 12
    drob1 = rasdel(drob1)
    drob2 = rasdel(drob2)
    chisl1 = int(drob1[0])
    znam1 = int(drob1[1])
    chisl2 = int(drob2[0])
    znam2 = int(drob2[1])
    rez = str(chisl1 * chisl2) + '/' + str(znam1 * znam2)
    rez = sokra(rez)
    
    if '/' in str(rez):
        
        symv = rez.find('/')
        l = len(rez)
        znam = rez[symv+1:l:1]
        chisl = rez[0:symv:1]
    
    if int(znam) == 1:
        
        return chisl
    
    return rez
```

```python
def delen(drob1, drob2):

    znam = 12
    drob1 = rasdel(drob1)
    drob2 = rasdel(drob2)
    chisl1 = int(drob1[0])
    znam1 = int(drob1[1])
    chisl2 = int(drob2[0])
    znam2 = int(drob2[1])
    rez = str(chisl1 * znam2) + '/' + str(znam1 * chisl2)
    rez = sokra(rez)
    
    if '/' in str(rez):
        
        symv = rez.find('/')
        l = len(rez)
        znam = rez[symv+1:l:1]
        chisl = rez[0:symv:1]
    
    if int(znam) == 1:
        
        return chisl
    
    return rez
```

```python
def vichdr(drobi):
    
    if '+' in drobi and '-' in drobi:
        
        symv = drobi.find('+')
        l = len(drobi)
        drobi = str(drobi[0:symv:1]) + str(drobi[symv+1:l:1])
    
    if '+' in drobi:
        
        drobi = drobi.replace(' ','')
        
        symv = drobi.find('+')
        l = len(drobi)
        drob2 = drobi[symv+1:l:1]
        drob1 = drobi[0:symv:1]
        
        return summa(drob1, drob2)
        
    elif '-' in drobi and '*' not in drobi:
        
        drobi = drobi.replace(' ','')
        
        symv = drobi.find('-')
        l = len(drobi)
        drob2 = drobi[symv+1:l:1]
        drob1 = drobi[0:symv:1]
        
        return ruznost(drob1, drob2)
        
    elif '*' in drobi:
        
        drobi = drobi.replace(' ','')
        
        symv = drobi.find('*')
        l = len(drobi)
        drob2 = drobi[symv+1:l:1]
        drob1 = drobi[0:symv:1]
        
        return umnoz(drob1, drob2)
    
    else:
        
        symv = drobi.find(' / ')
        l = len(drobi)
        drob2 = drobi[symv+2:l:1]
        drob1 = drobi[0:symv:1]
        
        return delen(drob1, drob2)
```

```python
def prplav(drob):
    
    drob1 = drob
    symv = drob1.find('(')
    l = len(drob1)
    drob1 = drob1[symv+1:l-1:1]
    
    cel = drob[0:symv:1]
    
    drob2 = rasdel(drob1)
    chisl = int(drob2[0])
    znam = int(drob2[1])
    
    kol2 = 0
    kol5 = 0
    
    while znam % 5 == 0 or znam % 2 == 0:
        
        if znam % 5 == 0:
            
            kol5 = kol5 + 1
            znam = znam // 5
            
        if znam % 2 == 0:
            
            kol2 = kol2 + 1
            znam = znam // 2
            
    if znam != 1:
        
        return 'Невозможно получить'
    
    ruznkol2 = kol2 - kol5
    ruznkol5 = kol5 - kol2
    
    while kol5 != kol2:
        
        if kol5 > kol2:
            
            kol2 = kol2 + 1
            
        else:
            
            kol5 = kol5 +1
            
    for i in range(kol2):
        
        znam = znam * 10
        
    if ruznkol2 > 0:
        
        chisl = 5 ** ruznkol2 * chisl
        
    if ruznkol5 > 0:
        
        chisl = 2 ** ruznkol5 * chisl
        
    schisl = str(chisl)

    while 1 + 1 == 2:
        
        if len(schisl) != kol5:
            
            schisl = '0' + schisl
            
        else:
            
            break
            
    rez = str(cel) + '.' + schisl
    
    return rez
```

```python
def skobki(vich):
    
    symvv = vich.rfind('(')

    dlinna = len(vich)
    
    for i in range(dlinna):

        if vich[i] == '(' and '0' <= vich[i-1] <= '9' and i - 1 >= 0:

            symvpl = vich.rfind('+', 0, i-1)
            symvmin = vich.rfind('-', 0, i-1)
            symvumn = vich.rfind('*', 0, i-1)
            symvdel = vich.rfind(' / ', 0, i-1)
            symvsk1 = vich.rfind('(', 0, i-1)
            symvsk2 = vich.rfind(')', 0, i-1)
            symvprob = vich.rfind(' ', 0, i-1)

            nyzsym = max(symvpl, symvmin, symvumn, symvdel, symvsk1, symvsk2, symvprob)

            if nyzsym == symvpl:
                
                granlev = symvpl
                
            if nyzsym == symvmin:
                
                granlev = symvmin
                
            if nyzsym == symvumn:
                
                granlev = symvumn
                
            if nyzsym == symvdel:
                
                granlev = symvdel
                
            if nyzsym == symvsk1:
                
                granlev = symvsk1
                
            if nyzsym == symvsk2:
                
                granlev = symvsk2
                
            if nyzsym == symvprob:
                
                granlev = symvprob
                
            granprav = vich.find(')', i)
            dlinaec = granprav - granlev + 1

            otvet = smesh(vich[granlev+1:granprav+1:1])
            vich = str(vich[:granlev+1]) + str(otvet) + str(vich[granprav+1:])

            vich = vich + ' ' * (dlinna - len(vich))
        
    while '(' in vich:

        granlev = vich.rfind('(')
        granprav = vich.find(')', granlev)
        otvet = bezskobki(vich[granlev+1:granprav:1])
        vich = str(vich[:granlev]) + str(otvet) + str(vich[granprav+1:])

    otvet = bezskobki(vich)
        
    return otvet
```
